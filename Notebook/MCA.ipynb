```python
import pandas as pd

# üßÆ Step 1 ‚Äî List your uploaded CSV filenames
# (Make sure these names match the files you uploaded)
file_names = [
    '/content/4dbe5667-7b6b-41d7-82af-211562424d9a_1e5c6d80580310d08ab5e80c84dfb7bf.csv',
    '/content/4dbe5667-7b6b-41d7-82af-211562424d9a_30948aedacec2d7d0b80ad794ed969c3.csv',
    '/content/4dbe5667-7b6b-41d7-82af-211562424d9a_3a23d1349c6682894f0ced239e3d91bb.csv',
    '/content/4dbe5667-7b6b-41d7-82af-211562424d9a_dff39c000aa72a24e9277a4913efcc74.csv',
    '/content/4dbe5667-7b6b-41d7-82af-211562424d9a_eef3fb3e2ad199156359edb2cf08fe31.csv'
]

# üßπ Step 2 ‚Äî Read, clean, and store all files in a list
dfs = []
for file in file_names:
    print(f"Processing ‚Üí {file}")
    df = pd.read_csv(file)

    # Standardize column names (lowercase + underscores)
    df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]

    # Remove duplicates based on CIN
    df.drop_duplicates(subset=['cin'], inplace=True)

    # Add a column for the state (derived from file name)
    df['state'] = file.split('.')[0].lower()

    # Fill missing values with 'NA' to avoid merge errors
    df.fillna('NA', inplace=True)

    dfs.append(df)

# üß© Step 3 ‚Äî Combine all dataframes into one master dataset
master_df = pd.concat(dfs, ignore_index=True)

print(f"‚úÖ Combined dataset created with {len(master_df)} records and {len(master_df.columns)} columns")

# üóÇÔ∏è Step 4 ‚Äî Save the merged file
master_df.to_csv('mca_master_cleaned.csv', index=False)
print("üìÅ Saved as: mca_master_cleaned.csv")

# üîç Step 5 ‚Äî Preview first few rows
master_df.head()
```

```python
import pandas as pd

# Load your master dataset
df_day1 = pd.read_csv('/content/mca_master_cleaned.csv', engine='python')

# Simulate a second-day dataset
df_day2 = df_day1.copy()

# Add a new company (New Incorporation)
new_company = {
    'cin': 'NEWCIN001',
    'companyname': 'AI Future Tech Pvt Ltd',
    'companystatus': 'Active',
    'authorizedcapital': 1000000,
    'paidupcapital': 500000,
    'companyregistrationdate_date': '2025-10-15',
    'registered_office_address': 'Pune, Maharashtra',
    'state': 'maharashtra'
}
df_day2 = pd.concat([df_day2, pd.DataFrame([new_company])], ignore_index=True)

# Update authorized capital for one existing company
# Ensure the index exists before attempting to update
if len(df_day2) > 5:
  df_day2.loc[5, 'authorizedcapital'] = df_day2.loc[5, 'authorizedcapital'] + 100000


# Remove one company (Deregistered)
# Ensure the index exists before attempting to drop
if len(df_day2) > 10:
  df_day2 = df_day2.drop(df_day2.index[10])


print("‚úÖ Simulated day1 & day2 datasets ready!")
```

```python
def detect_changes(df_old, df_new):
    changes = []
    merged = df_old.merge(df_new, on='cin', how='outer', suffixes=('_old', '_new'), indicator=True)

    for _, row in merged.iterrows():
        if row['_merge'] == 'left_only':
            changes.append(['Deregistered', row['cin'], None, None, None])
        elif row['_merge'] == 'right_only':
            changes.append(['New Incorporation', row['cin'], None, None, None])
        else:
            for col in df_old.columns:
                col_old = f"{col}_old"
                col_new = f"{col}_new"
                if col_old in row and col_new in row and row[col_old] != row[col_new]:
                    changes.append(['Updated', row['cin'], col, row[col_old], row[col_new]])

    return pd.DataFrame(changes, columns=['Change_Type', 'CIN', 'Field_Changed', 'Old_Value', 'New_Value'])

# Detect differences between the two days
change_log = detect_changes(df_day1, df_day2)

print("‚úÖ Changes detected:")
change_log.head()
```

```python
import pandas as pd

# Load datasets
change_log = pd.read_csv('change_log_day1.csv')
master_df = pd.read_csv('final_enriched_mca_master.csv')

def chat_with_mca(query):
    query = query.lower()

    if "new incorporation" in query or "new company" in query:
        return change_log[change_log['Change_Type'] == 'New Incorporation']

    elif "deregistered" in query or "strike off" in query:
        return change_log[change_log['Change_Type'] == 'Deregistered']

    elif "updated" in query:
        return change_log[change_log['Change_Type'] == 'Updated']

    elif "active companies" in query:
        return master_df[master_df['companystatus'].str.lower() == 'active'][['cin','companyname','state']].head(10)

    else:
        return "ü§ñ Sorry, I didn't understand your query. Try 'show new incorporations' or 'list deregistered companies'."

# Test the chatbot
chat_with_mca("show new incorporations in maharashtra")
```
